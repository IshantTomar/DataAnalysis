```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('dark_background')
data = np.arange(10)
data1 = np.random.rand(10)
```

```python
plt.plot(data, marker='v', label='data1')
```

```python
fig = plt.figure()
plot1 = fig.add_subplot(2,2, 1)
plot1.plot(data, linestyle='--', color='orange')
plot2 = fig.add_subplot(2,2, 2)
plot2.plot(data1, linestyle=':', color='lime')
plot3 = fig.add_subplot(2,2, 3)
plot3.plot(data, linestyle='--', color='orange', label='data')
plot3.plot(data1, linestyle=':', color='lime', label='data1')
```
