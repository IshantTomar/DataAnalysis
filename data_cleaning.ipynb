```python
tab = {
    'Product': {1: 'Apple', 2: 'Onion', 3: 'Pineapple', 4: 'Cabbage'},
    'Price': {1: 10, 2: 7, 3: 8, 4: 12},
    'Quantity': {1: 23, 2: 52, 3: 63, 4: 14}
}
df = pd.DataFrame(tab)
df
```

```python
product_to_category = {
    'apple': 'Fruit',
    'pineapple': 'Fruit',
    'onion': 'Vegetable',
    'cabbage': 'Vegetable',
}
lowercased = df['Product'].str.lower()
df['Category'] = lowercased.map(product_to_category)
df
```
