<h3><u>Some Index Methods and properties</u></h3>

```python
# append -> Concatenate with additional Index objects, producing a new Index
ind1 = pd.Index([1,2,3,4,5])
ind2 = pd.Index(['a','b','c','d',5])
ind1.append(ind2)
```

```python
# intersection -> Compute set intersection
ind1 = pd.Index([1,2,3,4,5])
ind2 = pd.Index(['a','b','c','d',5])
ind1.intersection(ind2)
```

```python
# isin -> Compute boolean array indicating whether each value is contained in the passed collection
ind1 = pd.Index([1,2,3,4,5])
ind2 = pd.Index(['a','b','c','d',5])
ind1.isin(ind2)
```

```python
# updates the existing data.
inplac = data_clean.dropna(inplace = True)  
data_clean
```

```python
new_data_clean = salary_data.copy()
# removes whole row if all value in row is nan.
new_data_clean.dropna(how='all')
```

<h3><u>Transforming Data Using a Function or Mapping</u></h3>

```python
tab1 = {
    'Product': {1: 'Apple', 2: 'Onion', 3: 'Pineapple', 4: 'Cabbage'},
    'Price': {1: 10, 2: 7, 3: 8, 4: 12},
    'Quantity': {1: 23, 2: 52, 3: 63, 4: 14}
}
new_df = pd.DataFrame(tab1)
new_df
```

```python
product_to_category = {
    'apple': 'Fruit',
    'pineapple': 'Fruit',
    'onion': 'Vegetable',
    'cabbage': 'Vegetable',
}
lowercased = new_df['Product'].str.lower()
new_df['Category'] = lowercased.map(product_to_category)
new_df
```

<h3><u>Discretization and Binning</u></h3>

```python
price = [20, 30, 52, 98, 71, 26, 47, 54]
bins = [18, 30, 50, 80, 100]
names = ['Cheap', 'Affordable', 'Normal', 'Expensive']
pd.cut(price, bins, labels=names)
```
