<h3><u>Some Index Methods and properties</u></h3>

```python
# append -> Concatenate with additional Index objects, producing a new Index
ind1 = pd.Index([1,2,3,4,5])
ind2 = pd.Index(['a','b','c','d',5])
ind1.append(ind2)
```

```python
# intersection -> Compute set intersection
ind1 = pd.Index([1,2,3,4,5])
ind2 = pd.Index(['a','b','c','d',5])
ind1.intersection(ind2)
```

```python
# isin -> Compute boolean array indicating whether each value is contained in the passed collection
ind1 = pd.Index([1,2,3,4,5])
ind2 = pd.Index(['a','b','c','d',5])
ind1.isin(ind2)
```

<h3><u>Reindexing</u></h3>

```python
ser = pd.Series([1,2,3,4,5], index = ['f','b','a','e','d'])
ser
```

```python
ser1 = ser.reindex(['a','b','c','d','e'])
ser1
```

<h3><u>Forward Fill Method</u></h3>

```python
obj1 = pd.Series(['one', 'three', 'two'], index=[2,4,6])
obj1
```

```python
obj1.reindex(range(7), method='ffill')
```

<h3><u>Reindexing with dataframe</u></h3>

```python
df1 = pd.DataFrame(np.arange(6).reshape((3, 2)), 
                     index=['a', 'd', 'f'],
                     columns=['Minecraft', 'Dota'])
df1
```

```python
df2 = df1.reindex(['a', 'b', 'c', 'd','e','f'])
df2
```

```python
# using column keyword
games = ['MLBB', 'Fortnite', 'Minecraft']
df1.reindex(columns=games)
```
