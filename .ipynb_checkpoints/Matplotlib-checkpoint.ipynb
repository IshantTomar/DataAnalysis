```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('dark_background')
data = np.arange(10)
data1 = np.random.rand(10)
```

```python
plt.plot(data, marker='v', label='data1')
```

```python
fig = plt.figure()
plot1 = fig.add_subplot(2,2, 1)
plot1.plot(data, linestyle='--', color='orange')
plot2 = fig.add_subplot(2,2, 2)
plot2.plot(data1, linestyle=':', color='lime')
plot3 = fig.add_subplot(2,2, 3)
plot3.plot(data, linestyle='--', color='orange', label='data')
plot3.plot(data1, linestyle=':', color='lime', label='data1')
# plt.savefig('./test.png', dpi=400, bbox_inches='tight')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(data, 'g', label = 'data')
ax.plot(data1, 'b--', label = 'data1')
ax.legend(loc = 'best')
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.4, 0.2), 0.4, 0.2, color='r', alpha=1)
circ = plt.Circle((0.4, 0.7), 0.18, color='g', alpha=0.3)
pgon = plt.Polygon([[0.1, 0.15], [0.35, 0.4], [0.1, 0.6]], color='b', alpha=0.6)
ax1.add_patch(rect)
ax1.add_patch(circ)
ax1.add_patch(pgon)
plt.savefig('./shapes_test.png', dpi=400, bbox_inches='tight')
```
